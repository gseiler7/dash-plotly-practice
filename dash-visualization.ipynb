```python
import dash
import dash_core_components as dcc
from dash import dcc,html
import plotly.express as px
import pandas as pd
# import pymssql

df = pd.read_csv('heart.csv')
df
```

```python
app = dash.Dash(__name__)
```

```python
df1 = df[['MaxHR','RestingBP']]
fig1 = px.scatter(df1, x='MaxHR',y='RestingBP',title='MaxHR vs RestingBP')
```

```python
df2 = df[['RestingBP','Cholesterol']]
fig2 = px.scatter(df2, x='RestingBP',y='Cholesterol',title='RestingBP vs Cholesterol')
```

```python
df3 = df['ChestPainType'].value_counts().keys().tolist()
df4 = df['ChestPainType'].value_counts().tolist()
fig3 = px.bar(y=df3,x=df4,orientation='h',title='Reported Chest Pain Type')
```

```python
app.layout = html.Div(children=[
    html.H1(children='Heart disease indicators'),

    html.Div(children='''
        RestingBP, MaxHR, and Cholesterol
    '''),

    dcc.Graph(
        id='MaxHR-vs-RestingBP',
        figure=fig1
    ),

    dcc.Graph(
        id='RestingBP-vs-Cholesterol',
        figure=fig2
    ),

    dcc.Graph(
        id='ChestPainType',
        figure=fig3
    )
])

if __name__ == '__main__':
    app.run_server(debug=True,use_reloader=False)
```
